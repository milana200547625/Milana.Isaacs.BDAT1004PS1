Question 1 C#
What data type is each of the following?
5
5.0
5 > 1
'5'
5 * 2
'5' * 2
'5' + '2'
5 / 2
5 % 2
{5, 2, 1}
5 == 3
Pi (the number)

Reference:
https://learn.microsoft.com/en-us/dotnet/api/system.collections.hashtable?view=net-7.0

```
int[] array1 = { 5, 2, 1 };
Hashtable inputs = new Hashtable(); //Create a new HashTable to store and print input and its datatype as key-value pair
inputs.Add("5", 5.GetType());
inputs.Add("5.0", 5.0.GetType());
inputs.Add("5 > 1", (5 > 1).GetType());
inputs.Add("'5'", '5'.GetType());
inputs.Add("5 * 2", (5 * 2).GetType());
inputs.Add("'5' * 2", ('5' * 2).GetType());
inputs.Add("'5' * '2'", ('5' * '2').GetType());
inputs.Add("5 / 2", (5 / 2).GetType());
inputs.Add("5 % 2", (5 % 2).GetType());
inputs.Add("{5, 2, 1}", array1.GetType());
inputs.Add("5 == 3", (5 == 3).GetType());
inputs.Add("Pi (the number)", 3.14.GetType());

foreach (DictionaryEntry de in inputs)
{
    Console.WriteLine("Data Type of {0} is {1}",de.Key,de.Value);
}
```

Question 3 C#
Implement function triangleArea(a,b,c) that takes as input the lengths of the 3
sides of a triangle and returns the area of the triangle. By Heron's formula, the area
of a triangle with side lengths a, b, and c is
s(s - a)(s -b)(s -c), where
s = (a +b + c) /2.
>>> triangleArea(2,2,2)
1.7320508075688772

```
using System;//Visual Studio syntax
namespace Triangle
{
    class findArea 
    {
        static void Main(string[] args) 
        {
            Console.Write("Enter length1 of Triangle: ");
            double l1 = double.Parse(Console.ReadLine());
            Console.Write("Enter length2 of Triangle: ");
            double l2 = double.Parse(Console.ReadLine());
            Console.Write("Enter length3 of Triangle: ");
            double l3 = double.Parse(Console.ReadLine());
            findArea ta = new findArea();
            double area = ta.triangleArea(l1, l2, l3);
            Console.WriteLine("Area of Triangle:" + area);
        }
        
        double triangleArea(double a, double b, double c)
        {
            double s = (a + b + c) / 2;
            double d = s * (s - a) * (s - b) * (s - c);
            double area = Math.Sqrt(d);
            return area;
        }
    }

}
```

Question 4 C#
Write a program in C# Sharp to separate odd and even integers in separate arrays.
Go to the editor
Test Data :
Input the number of elements to be stored in the array :5
Input 5 elements in the array :
element - 0 : 25
element - 1 : 47
element - 2 : 42
element - 3 : 56
element - 4 : 32
Expected Output:
The Even elements are:
42 56 32
The Odd elements are :
25 47

```
Console.Write("Input the number of elements to be stored in the array: ");
int size = int.Parse(Console.ReadLine());
Console.WriteLine("Input "+size+" elements in the array:");
int[] arrInt = new int[size];
int[] arrEven = new int[size] ;
int[] arrOdd = new int[size] ;
int j = 0, k = 0;
for (int i = 0; i < size; i++)
{
    Console.Write("element - "+ i +": ");
    arrInt[i] = int.Parse(Console.ReadLine());
    if (arrInt[i] % 2 == 0)
    {
        arrEven[j] = arrInt[i];
        j++;
    }
    else
    {
        arrOdd[k] = arrInt[i];
        k++;
    }
}
Console.Write("The Even elements are: ");
for (int i = 0; i < j; i++)
{
    Console.Write(arrEven[i]+" ");
}
Console.Write("\nThe Odd elements are: ");
for (int i = 0; i < k; i++)
{
    Console.Write(arrOdd[i]+" ");
}
```
